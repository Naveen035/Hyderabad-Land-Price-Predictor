```python
df = df[~(df["Price 2021"] == "2021(Rs)")]
df = df[~(df["Price 2021"] == "2021")]
```

```python
df['Growth'] = df['Growth'].replace("Constant", "0")
```

```python
def convert_sqyd_to_sqft(price_str):
    if "/sq. yd." in price_str:
        price_range = price_str.replace("/sq. yd.", "").strip()
        min_price, max_price = price_range.split(" - ")
        min_price_sqft = float(min_price.replace("Rs. ", "").replace(",", "").strip()) / 9
        max_price_sqft = float(max_price.replace("Rs. ", "").replace(",", "").strip()) / 9
        return f"Rs. {int(min_price_sqft):,} - {int(max_price_sqft):,}/sq. ft."
    else:
        return price_str
df["Price 2020"] = df["Price 2020"].apply(convert_sqyd_to_sqft)
df["Price 2021"] = df["Price 2021"].apply(convert_sqyd_to_sqft)
df.head(20)
```

```python
df["Price 2021"] = df["Price 2021"].str.replace("/sq. ft.","",regex = False)
df["Price 2020"] = df["Price 2020"].str.replace("/sq. ft.","",regex = False)
```

```python
df.head()
```

```python
df["Price 2021"] = df["Price 2021"].str.replace("Rs. ", "", regex=False)  
df["Price 2020"] = df["Price 2020"].str.replace("Rs. ", "", regex=False)  
```

```python
df[df["Price 2021"] == "2021(Rs)"]
```

```python
def process_price(value):
    if " - " in value:
        min_price, max_price = value.split(' - ')
        min_price = int(min_price.replace(',', '').strip())
        max_price = int(max_price.replace(',', '').strip())
        average_price = (min_price + max_price) / 2
        return average_price
    else:
        return int(value.replace(',', '').strip())

df['Price_in_2021 (sq.ft)'] = df['Price 2021'].apply(process_price)
df['Price_in_2020 (sq.ft)'] = df['Price 2020'].apply(process_price)
```
